```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_vnindex = pd.read_csv(r"C:\Users\Admin\Desktop\TANPHAT\hocotruong\Năm ba 2025-2026\HK1_A\Thu thập và tiền xử lý dữ liệu\Đồ_án_GDP\data\processed\VNINDEX_processed.csv")  
```

```python
df_vnindex.info()
```

```python
# Ép kiểu cột date về datetime
df_vnindex["date"] = pd.to_datetime(df_vnindex["date"], errors="coerce")
df_vnindex = df_vnindex[
        (df_vnindex['date'].dt.year >= 2010) & 
        (df_vnindex['date'].dt.year <= 2024)
    ].copy()
```

```python
df_vnindex.describe()
```

```python

plt.figure(figsize=(12,5))
plt.plot(df_vnindex['date'], df_vnindex['vnindex_close'], label='VNINDEX Close', color='blue')

# Tính giá trị trung bình
mean_value = df_vnindex['vnindex_close'].mean()

# Vẽ đường trung bình
plt.axhline(y=mean_value, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_value:.2f}')

plt.title('VNINDEX Daily Closing Price', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```
