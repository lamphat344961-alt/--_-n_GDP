```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# GDP tăng trưởng %
df_gdp_growth  = pd.read_csv(r"C:\Users\Admin\Desktop\TANPHAT\hocotruong\Năm ba 2025-2026\HK1_A\Thu thập và tiền xử lý dữ liệu\Đồ_án_GDP\data\processed\gdp_growth_VN.csv")  
df_gdp_growth.info()
# Chỉ số giá tiêu dùng tổng hợp
df_cpi_index   = pd.read_csv(r"C:\Users\Admin\Desktop\TANPHAT\hocotruong\Năm ba 2025-2026\HK1_A\Thu thập và tiền xử lý dữ liệu\Đồ_án_GDP\data\processed\cpi_index_VN.csv")        
df_cpi_index.info()
# Dòng vốn đầu tư trực tiếp
df_fdi_inflow  = pd.read_csv(r"C:\Users\Admin\Desktop\TANPHAT\hocotruong\Năm ba 2025-2026\HK1_A\Thu thập và tiền xử lý dữ liệu\Đồ_án_GDP\data\processed\fdi_inflow_VN.csv")  
df_fdi_inflow.info() 
```

```python
# Danh sách các DataFrame cần xử lý
dataframes = {
    "df_gdp_growth": df_gdp_growth,
    "df_cpi_index": df_cpi_index,
    "df_fdi_inflow": df_fdi_inflow
}

# Duyệt qua để ép kiểu và lọc dữ liệu
for name, df in dataframes.items():
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    df = df[(df['date'].dt.year >= 2010) & (df['date'].dt.year <= 2024)].copy()
    dataframes[name] = df  

# Gán ngược lại để dùng tiếp
df_gdp_growth = dataframes["df_gdp_growth"]
df_cpi_index = dataframes["df_cpi_index"]
df_fdi_inflow = dataframes["df_fdi_inflow"]
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_gdp_growth['date'], df_gdp_growth['gdp_percent'], marker='o', color='green', linewidth=2)
plt.title('Vietnam GDP Growth Rate (%)')
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color='red', linestyle='--', linewidth=1)
plt.show()
```

Giai đoạn 2011–2019: GDP duy trì ở mức 5.5%–7.5%, cho thấy nền kinh tế tăng trưởng ổn định.

2020–2021: Sụt giảm mạnh do tác động của COVID-19, chỉ còn khoảng 2.5%–3%.

2022: Phục hồi mạnh, vượt 8%, mức cao nhất trong nhiều năm.

2023–2025: Tăng trưởng chậm lại, dao động 5%–7%, phản ánh xu hướng ổn định nhưng chịu tác động từ suy giảm toàn cầu.

Nhận xét: Tăng trưởng GDP có xu hướng phục hồi tốt sau khủng hoảng, phản ánh năng lực thích ứng cao của nền kinh tế Việt Nam.

```python
plt.figure(figsize=(10,5))
plt.plot(df_cpi_index['date'], df_cpi_index['cpi_index'], color='orange', linewidth=2)
plt.title('Vietnam CPI Index (Base Year = 100)')
plt.xlabel('Year')
plt.ylabel('CPI Index')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

2011–2012: Lạm phát rất cao (9–18%), giai đoạn khủng hoảng giá cả và thắt chặt tiền tệ.

2013–2016: Giảm mạnh, xuống mức 1–5%, nhờ chính sách ổn định vĩ mô của Chính phủ.

2017–2025: CPI duy trì mức thấp và ổn định (2–5%), cho thấy lạm phát được kiểm soát tốt.

Nhận xét:
CPI giảm dần giúp duy trì sức mua, ổn định môi trường đầu tư — đây là một yếu tố quan trọng giúp FDI và GDP tăng trưởng bền vững.

```python
plt.figure(figsize=(10,5))
plt.plot(df_fdi_inflow['date'], df_fdi_inflow['fdi_inflow']/1e9, marker='o', color='green')
plt.title('Vietnam FDI Inflows (Billion USD)')
plt.xlabel('Year')
plt.ylabel('FDI (Billion USD)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

2011–2014: Mức FDI khá thấp (7–9 tỷ USD/năm).

2015–2019: Tăng mạnh liên tục, đạt 15–16 tỷ USD/năm nhờ môi trường đầu tư hấp dẫn, mở cửa kinh tế và ký kết các hiệp định thương mại.

2020–2022: Dù COVID-19, FDI vẫn duy trì ổn định, chỉ giảm nhẹ.

2023–2025: Tăng vọt trở lại, đạt hơn 20 tỷ USD, thể hiện niềm tin của nhà đầu tư nước ngoài.

Nhận xét:
FDI là động lực tăng trưởng chính cho Việt Nam — đóng góp vào xuất khẩu, việc làm và công nghệ. Dòng vốn tăng ổn định cho thấy Việt Nam vẫn là điểm đến đầu tư hấp dẫn ở châu Á.

```python
for df, col in [(df_gdp_growth, 'gdp_percent'), (df_cpi_index, 'cpi_index'), (df_fdi_inflow, 'fdi_inflow')]:
    df['norm'] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

plt.figure(figsize=(12,6))
plt.plot(df_gdp_growth['date'], df_gdp_growth['norm'], label='GDP Growth (normalized)', color='green')
plt.plot(df_cpi_index['date'], df_cpi_index['norm'], label='CPI Index (normalized)', color='red')
plt.plot(df_fdi_inflow['date'], df_fdi_inflow['norm'], label='FDI Inflow (normalized)', color='blue')

plt.title("So sánh tương đối: GDP Growth – CPI – FDI (Normalized 0–1)")
plt.xlabel("Năm")
plt.ylabel("Giá trị chuẩn hóa (0–1)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

Phân tích mối quan hệ:

Khi CPI giảm ổn định, FDI tăng mạnh → môi trường đầu tư ổn định khuyến khích dòng vốn.

GDP và FDI có xu hướng tăng song hành → đầu tư nước ngoài góp phần trực tiếp vào tăng trưởng kinh tế.

Sự sụt giảm GDP năm 2021 không làm FDI giảm mạnh, cho thấy niềm tin dài hạn của nhà đầu tư vẫn cao.
